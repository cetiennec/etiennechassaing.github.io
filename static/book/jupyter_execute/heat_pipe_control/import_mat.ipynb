```python
import pandas as pd
import matplotlib.pyplot as plt
# Load the CSV file
data = pd.read_csv("pipe_heater_data.csv")
```

```python

# Extract columns
t = data['Time'].values
Tinlet = data['Tinlet'].values
Toutlet = data['Toutlet'].values
Tambiant = data['Tambiant'].values
Command_Heater = data['HeaterCommand'].values

# Plot temperatures and power
plt.figure(figsize=(10, 5))

# Add horizontal lines for minimum and limit temperatures
plt.axhline(y=85, color='orange', linestyle='--', label="Minimum Temp (85°C)")
plt.axhline(y=95, color='red', linestyle='--', label="Limit Temp (95°C)")


plt.plot(t/60, Tinlet, label="$T_{inlet}$")
plt.plot(t/60, Toutlet, label="$T_{outlet}$")
plt.plot(t/60, 33*Command_Heater, label="Power")

# Add an arrow between 1 and 8 minutes with "delay" label
# Add a double-headed arrow between 1 and 8 minutes with "Delay" text above
plt.annotate('',
             xy=(8, 20), xytext=(0.5, 20),
             arrowprops=dict(arrowstyle="<->", lw=1.5, color='black'),
             ha='center', va='bottom')
plt.annotate('Delay',
             xy=(3, 22))

# Labels and legend
plt.xlabel("Time [min]")
plt.ylabel("Temperature [°C]/Power [kW]")
plt.title("Pipe Heater Temperatures for a 30kW power step on real system")
plt.legend(loc="lower right")
plt.grid(True)

# Save and show the plot
plt.savefig("images/initial_problem.png", dpi=300, bbox_inches='tight')
plt.show()
```
